```python
%load_ext autoreload
%autoreload 2

import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
import sys
import warnings
from pandas.errors import PerformanceWarning 
warnings.filterwarnings('ignore', category=PerformanceWarning)

root_path = '/home/jadli/Desktop/BDAI2/projects/Credit-risk/data/raw'
sys.path.append(os.path.abspath("../.."))

from src.model import train_model
from src.data import clean_feature_names

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)


# Charger les fichiers de données en utilisant le chemin racine
application_train = pd.read_csv(os.path.join(root_path, 'application_train.csv'))
application_test = pd.read_csv(os.path.join(root_path, 'application_test.csv'))
bureau = pd.read_csv(os.path.join(root_path, 'bureau.csv'))
bureau_balance = pd.read_csv(os.path.join(root_path, 'bureau_balance.csv'))
previous_application = pd.read_csv(os.path.join(root_path, 'previous_application.csv'))
credit_card_balance = pd.read_csv(os.path.join(root_path, 'credit_card_balance.csv'))
# HomeCredit_columns_description = pd.read_csv(os.path.join(root_path, 'HomeCredit_columns_description.csv'))
installments_payments = pd.read_csv(os.path.join(root_path, 'installments_payments.csv'))
POS_CASH_balance = pd.read_csv(os.path.join(root_path, 'POS_CASH_balance.csv'))
# sample_submission = pd.read_csv(os.path.join(root_path, 'sample_submission.csv'))
```

```python
from src.data import transform_bureau_tables, transform_application_table, transform_previous_and_pos_cash, transform_credit_card_balance, transform_installments_payments

df_application_train = transform_application_table(application_train)
application_test = transform_application_table(application_test)
bureau_merged = transform_bureau_tables(bureau, bureau_balance)
previous_application_and_pos_cash_merged = transform_previous_and_pos_cash(previous_application, POS_CASH_balance)
credit_card_merged = transform_credit_card_balance(credit_card_balance)
installments_payments_transformed = transform_installments_payments(installments_payments)


df_train_merged = df_application_train.merge(bureau_merged, on='SK_ID_CURR', how='left').merge(previous_application_and_pos_cash_merged, on='SK_ID_CURR', how='left').merge(credit_card_merged, on='SK_ID_CURR', how='left').merge(installments_payments_transformed, on='SK_ID_CURR', how='left')
df_test_merged = application_test.merge(bureau_merged, on='SK_ID_CURR', how='left').merge(previous_application_and_pos_cash_merged, on='SK_ID_CURR', how='left').merge(credit_card_merged, on='SK_ID_CURR', how='left').merge(installments_payments_transformed, on='SK_ID_CURR', how='left')

print(application_train.shape)
print(df_train_merged.shape)

print(application_test.shape)
print(df_test_merged.shape)

# Enregistrer le DataFrame dans un fichier CSV
df_train_merged.to_csv('/home/jadli/Desktop/BDAI2/projects/Credit-risk/data/train_data/df_train_merged.csv', index=False)
df_test_merged.to_csv('/home/jadli/Desktop/BDAI2/projects/Credit-risk/data/train_data/df_test_merged.csv', index=False)

print("DataFrames merged and saved successfully.")
```
